```python
import os
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
# Read files to memory and visualize some.
N = 100_000
DATA_PATH = '../../../data/fixed-map-vps-1v1/'
samples_df = pd.read_csv(os.path.join(DATA_PATH, 'samples.csv.gzip'), compression="gzip", nrows=N)
rewards_df = pd.read_csv(os.path.join(DATA_PATH, 'rewards.csv.gzip'), compression="gzip", nrows=N)
```

```python
df = pd.concat([samples_df, rewards_df], axis=1)
df
```

```python
corr_df = samples_df.corrwith(rewards_df['RETURN'])
# sn.heatmap(corr_df, annot=True)
# corr_df.plot(kind='barh')
# plt.show()
corr_df.dropna().sort_values(ascending=False)[:20]
```

```python
pd.DataFrame({'corr': corr_df, 'abs_corr': abs(corr_df)}).nlargest(500, columns='abs_corr')['corr'].to_dict()
pd.DataFrame({'corr': corr_df, 'abs_corr': abs(corr_df)}).sort_values('abs_corr', ascending=False)[:20]
```
